```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie = pd.read_csv('movies.csv')
```

```python
rating = pd.read_csv('ratings.csv')
```

```python
movie.isnull().sum()
movie.head()
```

```python
rating.isnull().sum()
```

```python
cv = CountVectorizer(tokenizer = lambda x: x.split('|'), analyzer='word')
```

## Rekomendasi Movies untuk Joko
#### Joko menyukai Superman vs. The Elite (2012

```python
x = cv.fit_transform(movie['genres'])
print(cv.get_feature_names())
print(x.toarray())
```

```python
similarityScore = cosine_similarity(x.toarray())
similarityScore
```

```python
Joko = movie.index[movie['title'] == 'Superman vs. The Elite (2012)'][0]
Joko
```

```python
JokoMovie = list(enumerate(similarityScore[Joko]))
JokoMovie
```

```python
JokoMovie = sorted(JokoMovie, key=lambda x: x[1], reverse=True)
JokoMovie
```

# Result for Joko

```python
RekomJoko = []
for i in JokoMovie[1:6]:
    RekomJoko.append(movie.iloc[i[0]])
RekomJoko = pd.DataFrame(RekomJoko)
RekomJoko
```
